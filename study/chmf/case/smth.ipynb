```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# анализ данных
ds = pd.read_csv(r'C:\Users\nshir\code\python\shit\chmf\case\data.txt')
# статистика по данным
print(ds.describe())
```

```python
# расположение точек в 2-Д
sns.pairplot(ds)
```

```python
x = ds['X'].values
z = ds['Z'].values
y = ds['Y'].values
```

```python
# расположение точек в 3-Д

fig = plt.figure(facecolor = "white")
ax = fig.gca(projection='3d')
ax.plot_trisurf(z, x, y, cmap=plt.cm.jet, linewidth=0.01)
```

```python
# этап 3
ds['Y'].describe()
# среднее находится не около нуля, поэтому делаем вывод, что в нашем случае деформация + поверхность
```

```python
# усредняем по MinMax
ds['Y'] = np.divide((ds['Y'] - ds['Y'].min()), (ds['Y'].max() - ds['Y'].min()))
ds['Y'].describe()
```
